```python
def seek_cross_secant(x_l, x_r, eps):
    if function(x_l) * function(x_r) > 0:
        return "none"
    
    x_prev_l = x_l - 50 * eps
    x_curr_l = x_l
    x_prev_r = x_r + 50 * eps
    x_curr_r = x_r
    
    list_x_new = []
    while abs(x_prev_l - x_curr_l) > eps and abs(x_prev_r - x_curr_r) > eps and abs(x_curr_r - x_curr_l) > eps:
        x_new = x_curr_r - (function(x_curr_r) * (x_curr_r - x_curr_l)) / (function(x_curr_r) - function(x_curr_l))
        
        if function(x_new) * function(x_curr_l) > 0:
            x_prev_l = x_curr_l
            x_curr_l = x_new
        else:
            x_prev_r = x_curr_r
            x_curr_r = x_new
        list_x_new.append(x_new)
    return list_x_new[-1:]
```

```python
ef derivative(x, y, i, j):
    if i == 0:
        if j == 0:
            return 4 * x**3 + 6 * y**2 * 2 * x
        if j == 1:
            return 6 * x**2 * 2 * y + 4 * y**3
    elif i == 1:
        if j == 0:
            return 3 * x**2 * y + y**3
        if j == 1:
            return x**3 + x * 3 * y**2
```

```python
def jakobi(x, y):
    matrix = np.zeros((2, 2))
    for i in range(matrix.shape[0]):
        for j in range(matrix.shape[1]):
            matrix[i][j] = derivative(x, y, i, j)
    return matrix
```
